```python
gb_df = df.groupby("State")
print(gb_df.head())
type(gb_df)
```

```python
#iterating over the states 

act_df = pd.DataFrame()

for name,group in gb_df:
    if act_df.empty:
        act_df = group.set_index('Year')[['Low.2018']].rename(columns= {'Low.2018':name})
    else:
        act_df = act_df.join(group.set_index('Year')[['Low.2018']].rename(columns= {'Low.2018':name}))
        
act_df.head()
        
```
